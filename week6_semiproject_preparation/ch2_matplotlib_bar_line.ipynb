## matplotlib line chart
가장 기본 그래프인 선 그래프를 그릴 수 있는 다양한 방법들을 알아봅니다.

```python
df_seoul = df[(df["전출지별"]=="서울특별시") & (df["전입지별"]!="서울특별시")]
```

### line chart 그리기 및 스타일링

경기도에서 서울로 전입한 사람 수를 line chart로 그려보면서 matplotlib의 기본 옵션들을 익혀보겠습니다. 먼저 plt.plot() 함수에 series를 전달하면 series의 index를 x축, series의 값들을 y축으로 하는 선 그래프를 그릴 수 있습니다. 그 외 옵션들은 다음과 같은 역할을 수행합니다.

- plt.figure: 도표 크기 지정
- plt.title: 도표 제목 지정
- plt.xticks: 도표 x축 각 틱별 라벨 지정
- plt.xlabel, plt.ylabel: x축 y축을 각각 설명하는 라벨 지정
- plt.legend: 도표 오른쪽 위에 선그래프를 설명하는 라벨 추가. 이 때, plot 함수 호출시에 라벨을 지정해주어야 함.

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager, rc

plt.rc("font", family="AppleGothic")
```

```python
from matplotlib import pyplot as plt

plt.style.use("ggplot")
plt.figure(figsize=(12, 6))
plt.title("서울 -> 경기 인구 이동")
plt.xlabel("기간")
plt.ylabel("이동 인구수")
plt.plot(df_seoul.loc["경기도"], label="서울 -> 경기", marker="o", markersize=3)
plt.xticks(size=8, rotation=45, ha="right")
plt.legend()
```

### 여러 그래프를 한 도표 안에 그리기
여러 데이터에 대해서 plt.plot을 호출하면 한 도표에 여러 그래프를 그릴 수 있습니다 

```python
df_seoul.index.values
```

```python
destinations = [
    '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '세종특별자치시'
]
```

```python
from matplotlib import pyplot as plt

plt.style.use("ggplot")
plt.figure(figsize=(12, 6))
plt.title("서울 -> 광역시 인구 이동")
plt.xlabel("기간")
plt.ylabel("이동 인구수")
for destination in destinations:
    plt.plot(df_seoul.loc[destination], label=f"서울 -> {destination}", marker="o", markersize=3)
plt.xticks(size=8, rotation=45, ha="right")
plt.legend()
```

### 화면 분할
matplotlib에서 여러개의 도표를 한번에 만들고 싶을 때가 있다. 그럴 때는 figure의 subplot 기능을 사용하여 axe 객체들을 만들고, 각 axe별로 그래프를 그려주면 됩니다.

```python
plt.style.use("ggplot")
fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sr_one = df_seoul.loc["경기도"]
sr_two = df_seoul.loc["인천광역시"]

ax1.plot(sr_one, label=f"서울 -> 경기도", marker="o", markersize=3)
ax2.plot(sr_two, label=f"서울 -> 인천광역시", marker="o", markersize=3, color="skyblue", markerfacecolor="blue")

ax1.set_xticklabels(sr_one.index, rotation=90, size=5)
ax2.set_xticklabels(sr_two.index, rotation=90, size=5)

ax1.set_title("서울 -> 경기")
ax1.legend()
ax2.set_title("서울 -> 인천")
ax2.legend()
```

## area graph

### stacked area graph
stacked 옵션을 True로 놓고 그래프를 그려보면 전체 전입자 수와 각 전입지별 비중을 한눈에 확인할 수 있습니다. area 그래프를 사용할 경우 대부분, 이렇게 stack 옵션을 적용하여 시각화를 합니다.

```python
transposed_df.plot(kind="area", stacked=True, alpha=0.5)
```

### Stacked bar graph
bargraph도 area와 마찬가지로 stack 옵션을 적용하여서 그래프를 그릴 수 있습니다. 주로 구성 비율의 변화 등을 보여줄 때 사용합니다.

```python
df_seoul.index
```

```python
destinations = [
    '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '세종특별자치시',
    '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도'
]
```

```python
col_years
```

```python
transposed_df = df_seoul.loc[destinations, col_years].transpose()
```

```python
transposed_df
```

```python
total = transposed_df.sum(axis=1)
```

```python
total
```

```python
ratio_df = transposed_df.div(total, axis=0)
```

```python
ratio_df
```

```python
first_row = ratio_df.iloc[0]
```

```python
first_row
```

```python
ratio_df
```

```python
ratio_df = ratio_df.sort_index(key=lambda x: first_row[x], axis=1, ascending=False)
```

```python
ratio_df
```

```python
ax = ratio_df.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 5),
)
plt.title("2010년대 서울 전출 인구 지역 비율")
ax.legend(loc="lower left", bbox_to_anchor=(-0.1, -0.3), ncol=8)
```
