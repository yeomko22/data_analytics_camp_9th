# ch10. two sample t-test

남학생과 여학생의 한달 평균 용돈의 차이, 40 ~ 50대와 20 ~ 30대의 특정 대통령 후보에 대한 지지율의 차이 등과 같이 자료를 통해 두 모집단에 대한 비교 분석을 하는 경우가 많습니다. 이 때 t-test를 활용할 수 있으며, 아주 강력합니다.

## two sample t-test

도시와 농촌, 두 모집단의 평균 소득에 차이가 있는지 알아보는 상황을 가정하겠습니다. 각각의 모집단으로부터 표본을 추출하고, 표본 평균과 표본 분산을 측정하면 아래와 같습니다.

두 모집단의 평균이 같지 않다는 가설을 검정하기 위해서 대립가설과 귀무 가설을 세워보겠습니다.

귀무 가설: 도시의 평균 소득은 농촌의 평균 소득과 같다.  
대립 가설: 도시의 평균 소득은 농촌의 평균 소득과 다르다.

$$H_{0}: \mu_{1}=\mu_{2}$$
$$H_{a}: \mu_{1}\neq\mu_{2}$$

이를 검정하기 위해서는 도시와 농촌의 소득 분포를 동시에 고려하여 검정을 해야 합니다.



같은 가설이지만 형태를 살짝 바꿔서 귀무 가설과 대립 가설을 세워보겠습니다.

귀무가설: 도시의 평균 소득과 농촌의 평균 소득의 차는 0이다.  
대립가설: 도시의 평균 소득과 농촌의 평균 소득의 차는 0이 아니다.  

$$H_{0}: \mu_{1}-\mu_{2}=0$$
$$H_{a}: \mu_{1}-\mu_{2}\neq0$$

이를 검정하기 위해서는 각 집단의 표본 평균의 차의 분포만 고려하면 되므로, 훨씬 단순해집니다. 즉, 두 모집단의 표본 평균의 차를 이용해서 두 모집단의 평균의 차에 대한 추정을 하는 것이 two sample t-test 입니다. 



## 등분산 여부에 따른 t 통계량 결정

t-test는 모집단의 분산을 모를 때 사용한다고 했습니다. 단일 표본 t-test와 마찬가지로 t 통계량을 구하고, t 분포표를 이용하여 검정하면 됩니다. 이 때, 두 모집단의 분산이 같은지, 다른지 여부에 따라서 계산해야하는 t 통계량의 수식이 달라집니다. 왜냐하면 두 모집단의 분산이 같으면 공통으로 활용할 수 있는 정보가 있고, 다를 경우엔 각각 따로 계산해야 하기 때문입니다.

### 두 모집단의 분산이 같을 경우 t 통계량

$$두\;모집단의\;분산이\;같을\;경우(\sigma_{1}^2=\sigma_{2}^2)$$

$$v=n_{1}+n_{2}-2$$

$$S_{p}^2(합동\;표본\;분산)=\frac{(n-1)S_{1}^2+(n_{2}-1)S_{2}^2}{n_{1}+n_{2}-2}$$

$$t=\frac{(\bar{X_{1}}-\bar{X_{2}}) - (\mu_{1}-\mu_{2})}{\sqrt{S_{p}^2(\frac{1}{n_{1}} + \frac{1}{n_{2}})}} \sim t_{v}$$


### 두 모집단의 분산이 다를 경우 t 통계량

$$두\;모집단의\;분산이\;같을\;경우(\sigma_{1}^2\neq\sigma_{2}^2)$$

$$v=\frac{(s_{1}^2/n_{1} + s_{2}^2/n_{2})^2}{\frac{(s_{1}^2/n_{1})^2}{n_{1}-1} + \frac{(s_{2}^2/n_{2})^2}{n_{2}-1}}$$

$$t=\frac{(\bar{X_{1}}-\bar{X_{2}})-(\mu_{1}-\mu_{2})}{\sqrt{\frac{S_{1}^2}{n_{1}} + \frac{S_{2}^2}{n_{2}}}}\sim t_{v}$$

수식만 봐도 현기증이 나지만 전혀 겁먹을 필요가 없습니다. 여기서부터는 이제 컴퓨터에게 계산을 맞기면 됩니다. 다만, 기억해야할 것은 두 모집단의 분산이 같을 때와 다를 때 t 통계량을 계산하는 수식이 달라지는 부분입니다. 그런데 앞서서 t-test는 모집단의 분산을 모를 때, 사용한다고 했습니다. 그런데 각 모집단의 분산이 같은지 다른지를 어떻게 알 수 있을까요?

### 등분산 검정을 통한 t 통계량 결정
두 모집단의 분산이 같은지, 다른지를 알기 위해서는 또 다른 통계적 가설 검정 방법인 levene test를 사용할 수 있습니다. levene test의 귀무 가설은 "두 집단의 모분산은 같다"입니다. leven test 결과 p-value 값이 유의 수준보다 높다면 등분산 t 통계량을 계산하면 되고, p-value 값이 유의 수준보다 낮으면 이분산 t 통계량을 계산하면 됩니다.

## p-value 계산 및 통계적 의사 결정
두 집단의 평균이 같은지 여부를 비교하는 t-test를 수행하기 위해서는 등분산 검정 중 하나인 levene test를 수행합니다.  그 결과 도출된 p-value를 이용해 t 통계량을 선택하여 계산한 뒤, t-test의 p-value를 계산하면 됩니다. p-value가 유의 수준보다 낮다면 "두 집단의 평균의 차는 0이다." 라는 귀무 가설을 기각할 수 있습니다.

## 예제 1.

*Q. 특정 학급 학생들의 수학 시험 점수 데이터 셋이 주어졌습니다. 시험 점수는 각각 G1, G2, G3 세 컬럼이며 시간 순서대로 친 시험입니다. 남학생과 여학생의 점수의 평균이 같은지 통계적으로 검정해보세요.*

```python
import pandas as pd

df = pd.read_csv("./data/student-mat.csv")
```

```python
df
```

통계적 가설 검정을 하기 전에 우선 성별에 따른 G1, G2, G3 평균 점수를 확인해보겠습니다.

```python
df.groupby(["sex"])[["G1", "G2", "G3"]].mean()
```

결정론적인 관점에서보면 단순히 남학생이 여학생보다 수학을 더 잘한다고 결론지을 수도 있습니다. 하지만 이제 우리는 t-test를 배웠으니, 통계학적인 관점에서 남학생이 우연히 여학생 보다 수학 시험을 더 잘쳤을 확률을 계산해보고, 이 차이가 유의미한지 검정해보겠습니다.

### 가설 설정
모든 시험에서 남학생이 여학생보다 평균 점수가 높으므로 단측 검정으로 가설을 세워보겠습니다.

귀무 가설: 남학생과 여학생의 평균 점수의 차는 0이다.  
대립 가설: 남학생과 여학생의 평균 점수의 차는 0보다 크다. (단측 검정 선택)

$$H_{0}: \mu_{1}-\mu_{2}=0$$
$$H_{a}: \mu_{1}-\mu_{2}>0$$

### 등분산 검정
t-test를 수행하기 전에 levene test를 이용하여 두 모집단의 분산이 같은지 여부를 검정하겠습니다.

```python
from scipy import stats

test_names = ["G1", "G2", "G3"]

male_scores = df[df['sex'] == 'M'][['G1', 'G2', 'G3']]
female_scores = df[df['sex'] == 'F'][['G1', 'G2', 'G3']]

for test_name in test_names:
    _, p_value_levene = stats.levene(male_scores[test_name], female_scores[test_name])
    if p_value_levene > 0.05:
        print(f"{test_name} p-value: {p_value_levene}, 등분산 가정 만족")
    else:
        print(f"{test_name} p-value: {p_value_levene}, 이분산 가정 만족")
```

### t-test 수행
모든 시험 점수들에 대해서 등분산 가정을 만족하므로, 등분산 two sample t-test를 수행해보겠습니다.

```python
t_statistic, p_value = stats.ttest_ind(
    a=male_scores,
    b=female_scores,
    alternative="greater",
    equal_var=False
)
```

```python
print(f"p-value: {p_value}")
print(f"귀무 가설 기각: {p_value < 0.05}")
```

G1, G2, G3 세 시험에서 모두 p-value가 유의수준 보다 낮아서 귀무 가설을 기각하고 대립가설을 채택할 수 있게 되었습니다. 즉, 남학생의 평균 수학 점수는 여학생의 수학 점수보다 통계적으로 유의하게 높다고 결론 내릴 수 있습니다.
