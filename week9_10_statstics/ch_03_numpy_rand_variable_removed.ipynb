# ch3. numpy 난수 생성

numpy를 이용하여 난수를 생성할 수 있습니다. 그냥 난수를 생성할 수도 있지만 확률 분포를 지정해서 난수를 생성할 수도 있습니다. 이번 챕터에서는 numpy가 제공하는 난수 생성 기능을 사용해보면서 확률 분포 개념을 익혀보겠습니다. 

## 확률 변수와 확률 분포 
우리가 생성하고자 하는 random 난수는 특정한 확률에 의해서 값이 결정됩니다. 이를 확률 변수라고 부릅니다. 확률변수가 취할 수 있는 값을 확률에 대응시켜주는 함수를 확률 분포(probability distribution)라고 합니다. 더 쉽게 말하면 확률 변수들이 어떠한 형태로 놓여있을까, 어떻게 분포해 있을까를 나타내주는 함수입니다. 우리가 여러번 들어본 정규 분포가 가장 많이 쓰이는 확률 분포입니다.

### Uniform Distribution
uniform distribution은 확률 변수가 값을 가질 수 있는 모든 구간에서 동일한 확률을 갖는 것을 의미합니다. 가령, a와 b 구간에서 무작위의 수를 갖는 확률 분포는 아래와 갖습니다. 모든 경우에 동일한 확률 값을 가지므로 선분 형태를 보입니다. 

### Normal Distribution
normal distribution(정규 분포)는 평균을 중심으로 종 모양의 형태를 갖습니다. 정규분포의 형태는 평균과 표준 편차에 의해서 결정됩니다. 평균을 중심으로 좌우 대칭인 종모양을 가지고, 표준편차가 클 수록 완만하게 퍼져있는 형태를 갖습니다. 정규 분포 중에서도 평균이 0, 표준 편차가 1인 정규 분포를 표준 정규 분포라고 부릅니다.

## Numpy를 이용한 난수 생성
numpy random 모듈을 이용하면 난수를 생성할 수 있습니다. 이 때, 특정 확률 분포에 따라서 난수를 생성할 수 있습니다.

### np.random.random
np.random.random을 호출하면 파라미터로 전달한 개수만큼 0과 1 사이 구간에서 uniform distribution에 따라서 난수를 생성합니다. 직접 생성해보고, 결과를 히스토그램으로 그려보겠습니다.

난수를 많이 생성할 수록 모든 구간에서 균일한 개수의 난수가 생성되는 것을 확인할 수 있습니다. 

### np.random.randn

np.random.randn을 호출하면 파라미터로 전달받은 개수만큼 평균이 0, 표준 편차가 1인 표준 정규 분포에 따라서 난수를 생성합니다. 0을 중심으로 종모양으로 퍼져있기 때문에 난수 값이 음수가 될 수 도 있습니다.

난수를 많이 생성할 수록 점점 더 종모양을 닮아가는 것을 확인 할 수 있습니다.
