## t-test 개념

앞서 배웠던 z-test는 모집단이 정규분포를 따르고, 모분산을 알 때, 표본 평균의 값으로 모평균 값을 추정하는 기법이었습니다. 그런데 현실에서 대부분의 경우 모분산을 알 수 없습니다. 이런 경우 사용할 수 있는 기법이 t-test 입니다. 실제 현업에서도 자주 사용하는 기법입니다.

t-test인 이유는 t-test 논문을 발표한 저자가 당시 Student 라는 필명을 썼기 때문이라는 설이 있습니다. 그래서 student의 t를 따서 t-test, 여기서 사용하는 통계량을 t-statistic, 분포를 t 분포라고 부릅니다.

 t-test는 주로 아래 같은 상황에 적용됩니다.

- 모분산을 모를 때, 모평균이 특정 값인지 검정
    -  K 젤리 표본 100개의 표본 평균이 68.5g 일 때, 모평균이 70g인지 검정
- 두 집단의 평균이 같은지, 다른지 검정
    - A 대학 남학생들의 평균 키와 B 대학 남학생들의 평균 키가 같은지 다른지 검정
- 한 집단의 이전 평균과 이후 평균이 같은지 다른지 검정. 
    - 동일한 집단의 작년도 시험 점수와 다음 연도 시험 점수가 같은지 다른지 검정

### t 통계량
t-test를 수행하기 위해서는 먼저 t 값을 알아야 합니다. 먼저 중심 극한 정리에 의해서 표본 평균은 아래와 같은 정규 분포를 따릅니다.
$$\bar{X}\sim N(\mu,\frac{\sigma^2}{n})$$


그런데 우리는 현재 모분산을 모르는 상태입니다. 때문에 모분산을 표본 표준 편차로 대체한 t 통계량을 사용해야 합니다. 그리고 이렇게 구한 t 통계량은 t 분포를 따릅니다.

$$\sigma\rightarrow S$$

$$t=\frac{\bar{X}-{\mu}}{S/\sqrt{n}}$$

### t 분포
t 값의 확률 분포입니다. 자유도에 의해서 형태가 결정되며, 자유도의 크기가 커질 수록 표준 정규 분포에 근사합니다. 이를 직관적으로 이해하면 모분산을 모르더라도, 표본을 아주 많이 뽑으면 t 값은 표준 정규 분포를 따른다 정도로 이해할 수 있습니다.



표준 정규 분포표처럼 t 분포도 자유도에 따라서 미리 구간별 확률값을 계산해놓은 t 분포 테이블이 있습니다. 따라서 귀무 가설과 대립 가설을 세우고, 표본으로 부터 t 값을 계산한 다음, t 분포표를 이용해서 p-value를 계산하여 유의수준 보다 높은지 낮은지를 검정하는 것이 t-test 입니다.

## t-test 예제

*Q. 건강한 성인의 콜레스테롤 수치는 220 이하라고 한다. 콜레스테롤이 높은 사람은 심장마비가 발생할 가능성이 더 높은 것으로 알려져있다. 이를 확인해보기 위해 30명의 심장마비 환자를 조사하여 평균 콜레스테롤이 231이며 표준 편차가 20인 것을 알아냈습니다. 심장마비 환자들의 평균 콜레스테롤은 정상인의 콜레스테롤보다 높은 수치라고 결론 내릴 수 있는지 검정하시오*

1. 귀무 가설, 대립가설, 유의 수준 설정
2. 표본으로부터 t 통계량 계산
3. t 분포표를 보고 기각역 확인 및 t 통계량과 비교
4. 통계적 의사 결정

### 1. 귀무 가설, 대립 가설, 유의 수준 설정

귀무 가설: 심장마비 환자들의 평균 콜레스테롤 수치는 정상인과 같다.  
대립 가설: 심장마비 환자들의 평균 콜레스테롤 수치는 정상인보다 높다.  
유의 수준: 0.05

$$H_{0}: \mu=220$$
$$H_{a}: \mu>220$$
$$\alpha=0.05$$

### 2. t 통계량 계산

$$t=\frac{\bar{X}-\mu}{s/\sqrt{n}}=\frac{231-220}{20/\sqrt{30}}=3.0125$$

### 3. t 분포표 보고 기각역 확인 및 t 통계량과 비교

t 통계량을 구했으니 이제 t 분포표를 보고 p-value를 계산할 일만 남았습니다. 그런데 t 분포표는 아래와 같이 생겼습니다.


앞서 t 분포는 형태가 자유도에 의해서 결정된다고 했습니다. 우리가 컴퓨터라면 확률 밀도 함수를 직접 적분해서 구간 확률 값을 구하면 되겠지만, 지금처럼 손으로 풀 때는 무척 어렵습니다. 또한 모든 자유도에 대해서 구간 확률을 계산해서 제공할 경우, 양이 너무 방대합니다. 때문에 단측 검정인지 양측 검정인지, 자유도는 어떤지, 유의 수준은 어떤지에 따라서 주요 구간의 t 값을 미리 계산해 놓는 것입니다. 이를 기각역(Critical Region)이라고 부릅니다. 

우리 예제는 자유도는 30 - 1인 29, 유의 수준은 0.05인 단측 검정이었습니다. t 분포표를 통해 확인한 기각역은 1.699 였습니다. 이를 그래프로 그려보겠습니다.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 29
critical_region = 1.699
x = np.linspace(-5, 5, 500)
pdf = t.pdf(x, df)

plt.plot(x, pdf)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title(f"t-distribution df: {df}")
plt.fill_between(x, pdf, where=(x >= critical_region), color="red")
plt.legend()
```

자유도가 29인 t 분포에서 기각역인 1.699 이상인 영역의 확률 값은 0.05입니다. 즉, 가설 검정을 위해서 구한 t 통계량이 1.699 보다 크면 유의 수준인 0.05보다 작아져서 귀무 가설을 기각할 수 있게 됩니다. 

### 4. 통계적 의사결정

앞서 구했던 t 값은 3.0125로 기각역인 1.6999 보다 크므로 귀무 가설을 기각할 수 있습니다. 그러므로 심장병 환자들의 평균 콜레스테롤 수치는 일반인보다 높습니다.

$$기각역:\;t>1.6999\;이면\;H_{0}\;기각$$

## 연습 문제
*Q: A 라는  사람이 매일 아침 베이커리에 들려서 식빵을 구매합니다. 이 식빵의 무게는 500g 이라고 표시되어 있습니다. 하지만 이 A 라는 사람은 자신이 구매하는 빵이 500g 이 맞는지 의구심이 들기 시작합니다. 그래서 일주일 동안 식빵을 사온 뒤 매일 무게를 기록하였고 아래와 같습니다. 빵 무게가 500g 보다 작은지 통계적으로 가설 검정을 해보세요.*

- t 분포표를 보고 기각역을 사용하여 가설 검정을 해보세요.
- t 분포표 계산기를 이용하여 p-value를 구하고, 이를 이용해 가설 검정을 해보세요.
    - https://homepage.divms.uiowa.edu/~mbognar/applets/t.html

```python
import numpy as np

weights = np.array([495, 496, 502, 494, 499, 501, 493])
print("평균 무게: ", weights.mean())
print("표준 편차: ", weights.std(ddof=1))
```

1. 가설

귀무가설: 식빵의 무게의 평균은 500g이다.  
대립가설: 식빵의 무게의 평균은 500g이 아니다. (양측 검정)

2. t-value 계산  
(497 - 500) / (3.5 / sqrt(7)) = -2.267

3. t 분포표를 이용한 기각역 확인  
t > 2.447 or t < -2.447 이어야 귀무가설 기각이 가능

4. 통계적 의사 결정  
귀무 가설을 기각할 수 없다. 왜냐? t 값이 기각역에 속하지 못한다.

1. 가설

귀무가설: 식빵의 무게의 평균은 500g이다.  
대립가설: 식빵의 무게의 평균은 500g 보다 작다. (단측 검정)

2. t-value 계산  
(497 - 500) / (3.5 / sqrt(7)) = -2.267

3. t 분포표를 이용한 기각역 확인  
t > 1.943 or t < -1.943 이어야 귀무가설 기각이 가능

4. 통계적 의사 결정  
귀무 가설 기각이 가능하다. 즉, 식빵 무게는 500g 보다 작다.
