series index로 접근 가능함과 동시에 순서 index로도 접근이 가능하다.

# Pands DataFrame
Series가 index를 가지는 1차원 배열이었다면 DataFrame은 행과 열로 구성된 2차원 테이블입니다. pandas를 이용한 데이터 분석을 하면서 직접 dataframe을 한땀 한땀 만들 일은 드물지만, dataframe에 친숙도를 높이기 위해서 여러 방법으로 데이터 프레임을 만들어보겠습니다.

## DataFrame 만들어보기
### python list로부터 DataFrame 만들기
일일이 컬럼을 지정한 다음, 값에 해당하는 list를 전달하여 dataframe을 만들 수 있습니다.

index와 values를 따로 따로 list로 만들어서 data frame을 만들 수도 있습니다.

### python dictionary를 이용하여 dataframe 만들기
python 딕셔너리를 이용해서도 dataframe을 만들 수 있습니다.

### 특정 컬럼을 index로 지정하기
dataframe에서 행을 구분하는 값을 index라고 부릅니다. 앞서 만든 데이터 프레임은 생성시에 아무런 인덱스를 지정하지 않았기 때문에 0, 1, 2, 3 처럼 자동으로 증가하는 숫자 인덱스를 갖게 됩니다.

보통 데이터 프레임은 증가하는 숫자 index를 갖습니다만, 특정 컬럼의 값을 index로 사용할 수도 있습니다. 이 때, df.set_index 함수를 사용할 수 있습니다. 이 때, index로 지정하는 컬럼이 중복된 값을 포함하여도 가능합니다.

## DataFrame 값들에 접근
dataframe에서 값들을 조회하기 위해서는 크게 세 가지 방식이 있습니다. 특정 열 선택은 column의 이름을 이용하고, 특정 행을 선택하기 위해서는 index 혹은 행 번호를 이용합니다. 하나씩 알아보겠습니다. 

- colum 이름으로 조회하기
- df.loc으로 특정 행 데이터 조회하기
- df.iloc으로 i번째 행 데이터 조회하기

### column을 이용한 데이터 프레임 조회
df["컬럼명"] 혹은 df[["컬럼명 1", "컬럼명 2"]] 이런식으로 데이터 프레임의 특정 컬럼 값들을 조회할 수 있습니다.

### index를 이용한 컬럼값 조회
df.loc을 이용하면 index를 이용해서 데이터들을 조회할 수 있습니다. 특정 컬럼의 값을 index로 지정했을 때 많이 사용합니다.

### iloc을 이용한 행 접근
df.iloc을 이용하면 i번째 행에 접근할 수 있습니다. 

iloc을 사용하면 for문을 돌면서 dataframe의 각 행들을 출력할 수 있습니다. 행 단위 작업을 수행할 때, 유용합니다.
